```python
#문제 1
#성별 연령별 인구 데이터 꺾은선 그래프로 표현
import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')  
for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i]))   # 3 ~ 103번까지 데이터 더하기
            f.append(int(row[i+103]))  # 106 ~ 206번까지 데이터 더하기
        break
        
plt.plot(m, label = "Male")
plt.plot(f, label = "Female")
plt.legend()
plt.show()
```

```python
#문제 2
#연령별 남성 인구, 여성 인구 값 막대그래프로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
result = []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')  
for row in data :
    if name in row[0] :
        for i in range(3, 104) :
           result.append(int(row[i]) - int(row[i+103]))
        break
        
plt.bar(range(101), result)
plt.show()
```

```python
#제주도의 연령대별 성별 비율을 산점도로 표현하시오
import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')  
for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i]))   # 3 ~ 103번까지 데이터 더하기
            f.append(int(row[i+103]))  # 106 ~ 206번까지 데이터 더하기
        break
        
plt.style.use('ggplot')
plt.scatter(m,f)
plt.show()
```

```python
#제주도의 연령대별 성별 비율을 산점도로 표현하시오(y=x 형태의 추세선 추가)
import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')  
for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i]))   # 3 ~ 103번까지 데이터 더하기
            f.append(int(row[i+103]))  # 106 ~ 206번까지 데이터 더하기
        break
        
plt.style.use('ggplot')
plt.scatter(m,f, c=range(101), alpha=0.5, cmap='jet')
plt.plot(range(max(m)),range(max(m)), 'g')
plt.colorbar()
plt.show()
```

```python
#제주도의 연령대별 성별 비율을 산점도로 표현하시오(y=x 형태의 추세선 추가)
#연령대별 성별 비율 산점도로 표현하기(원의 크기는 인구수의 제곱근)
import csv
import matplotlib.pyplot as plt
import math
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')  
for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i]))   # 3 ~ 103번까지 데이터 더하기
            f.append(int(row[i+103]))# 106 ~ 206번까지 데이터 더하기
            size.append(math.sqrt(int(row[i])+int(row[i+103])))
        break
        
plt.style.use('ggplot')
plt.rc('font',family="Malgun Gothic")
plt.figure(figsize=(10,5),dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.scatter(m,f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.plot(range(max(m)),range(max(m)), 'g')
plt.colorbar()
plt.show()
```

```python
#헤더 제외하고 map 함수를 활용한 데이터 타입 바꾸기(문자열 -> 정수)
import csv

f=open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
for row in data:
    row[4:] = map(int, row[4:])
    print(row)
```

```python
#문제7
#아침 7시 승차 데이터 개수 및 인원수 출력하기 (result리스트에 저장하여 출력)
import csv

f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data:
    row[4:] = map(int, row[4:])
    result.append(row[10])
print(len(result))
print(result)
```

```python
#문제8
#아침 7시 승차 데이터 막대 그래프로 표현하기
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data:
    row[4:] = map(int, row[4:])
    result.append(row[10])

plt.style.use('ggplot')
plt.bar(range(len(result)), result)
plt.show()
```

```python
#문제9
#아침 7시 승차 데이터 오름차순으로 정렬
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data:
    row[4:] = map(int, row[4:])
    result.append(row[10])

result.sort()
plt.style.use('ggplot')
plt.bar(range(len(result)), result)
plt.show()
```

```python
#문제10
#아침 7~9시 승차 인원 더해서 오름차순으로 정렬
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data:
    row[4:] = map(int, row[4:])
    result.append(row[10]+row[12]+row[14]) #or result.append(sum(row[10:15:2]))

result.sort()
plt.style.use('ggplot')
plt.bar(range(len(result)), result)
plt.show()
```

```python
#문제11
#아침 7~9시 승차 인원 최대 역과 인원 수 찾기
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
max = 0
max_station = ''

for row in data:
    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > max:
        max = sum(row[10:15:2])
        max_station = row[3]+"("+row[1]+ ")"
    
print(max_station, max)
```

```python
#문제12
#아침 7~9시 하차 인원 최대 역과 인원 수 찾기
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
max = 0
max_station = ''

for row in data:
    row[4:] = map(int, row[4:])
    if sum(row[11:16:2]) > max:
        max = sum(row[11:16:2])
        max_station = row[3]+"("+row[1]+ ")"
    
print(max_station, max)
```

```python
#문제12
#밤 11시에 사람들이 가장 많이 타고 내리는 역 찾기
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
max = 0
max_station = ''

time = int(input("몇 시의 승차 인원이 가장 많은 역이 궁금하세요? : "))

for row in data:
    row[4:] = map(int, row[4:])
    a = row[4+(time-4)*2]
    if a > max:
        max = a
        max_station = row[3]+"("+row[1]+ ")"
    
print(max_station, max)
```
