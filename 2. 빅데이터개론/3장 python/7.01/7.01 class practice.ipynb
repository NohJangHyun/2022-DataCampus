```python
#MSE 오차 계산하기 - 예측값과 실제값의 오차
import numpy as np

ypred = np.array([1, 0, 0, 0, 0])
y = np.array([0, 1, 0, 0, 0])

ydif = ypred-y
MSE = np.sum(ydif**2)/len(ypred)
# MSE = (1/len(ypred))*np.sum(np.square(ypred-y))
print(MSE)
```

```python
#데이터 정의
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'name':['Kim','Lee','Park','Choi','Hong','Chung','Jang'],
    'age':[22,26,78,17,46,32,21],
    'city':['Seoul','Busan','Seoul','Busan','Seoul','Daejun','Daejun'],
    'children':[2,3,0,1,3,4,3],
    'pets':[0,1,0,2,2,0,3]
})
```

```python
#그래프 그리기
df.plot(kind="line", x="name", y="pets", color="red")
plt.show()
```

```python
#중첩 차트 그리기
ax = plt.gca() #gca(): get current axis

df.plot(kind='line',x='name',y='children',ax=ax)
df.plot(kind='line',x='name',y='pets', color='red', ax=ax)

plt.show()
```

```python
#막대 그래프 그리기
df.plot(kind='bar',x='name',y='age')
plt.show()
```

```python
#산포도 그리기
df.plot(kind='scatter',x='children',y='pets',color='red')
plt.show()
```

```python
#city로 그룹핑한 후 유일한 이름의 개수를 계산하여 막대 그래프로 그리기
df.groupby('city')['name'].nunique().plot(kind='bar')
plt.show()
```
