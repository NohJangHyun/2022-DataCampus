Series
pandas의 기본 객체 두 종류 중 하나
- One-dimensional : 1차원
- ndarray : Numpy 기본 객체 타입
- axis labels : 라벨 축(고유 값 가짐)
- including time seris : 시계열 포함(시간이 라벨 축일 수 있음)
- 라벨 축을 가진 1차원 ndarray(시계열 데이터 포함 - 순차적으로 정해진 데이터 셋의 집합)

DataFrame
padas의 기본 객체 두 종류 중 다른 하나
- Two-dimensional : 2차원
- size-mutable : 크기 변경 가능한 
- potentially heterogeneous : 잠재적으로 이종
- tabular data : 행(row), 열(column) 형태의 테이블 구조 데이터
- 크기가 유연하고, 서로 다른 자료형도 담을 수 있는 2차원 테이블 구조


mutable immutable 차이
mutable : 변경이 가능한 객체로 최초 생성 이후에 자유롭게 값의 변경, 추가 ,삭제 가능 (int, float, list, 사용자 정의 클래스)
immutable : 변경이 불가능한 객체로 최초 생성 이후에 값을 변경할 수 없음(tuple, string, dictionary - key, 내장 타입인 숫자, bool 등)

Indexing / Slicing
기존 Python, Numpy와 pandas의 Indexing / Slicing 차이점은 Index 정보가 정수가 아니어도 됨
- 기존 Python, Numpy : 인스턴스(정수)
- pandas의 Series, DataFrame : 인스턴스(index 정보)

Slicing에서 마지막 값에 대한 처리
- 기존 Pythonm Numpy [3:5]
  - 3부터 4까지(5 미포함)
- pandas의 Series, DataFrame [3:5]
  - 순번일 경우, 3부터 4까지(5 미포함)
  - index 정보일 경우, 3부터 5까지(5 포함)
- loc, iloc을 이용하여 복잡한 Indexing / Slicing 가능

```python
#객체 Series에서 Indexing / Slicing
ss = pd.Series(np.arange(4.), index=['a', 'b', 'c', 'd'])

print(ss['b'])
print(ss[1])
print(ss[:2])

print(ss[['b', 'a', 'd']])
print(ss[[1, 3]])

print(ss[ss < 2])
```

```python
print(ss['b':'c'])

ss['b':'c'] = 5
print(ss)
```

```python
#정수 색인 - 에러 발생
#기존 python 문법에서 사용한 음수 형태의 indexing / slicing기법은 index 정보에 따라 사용 가능/불가능함
#인스턴스 ser는 index 정보가 없으므로 자동으로 0부터 시작하는 index 정보가 생성됨
#-1 이라는 index 정보를 탐색하고 오류 발생 : KeyError
ser = pd.Series(np.arange(3.))
print(ser[-1])
```

```python
#비슷한 경우 : index 정보가 정수일 때 발생
#ser의 index [0, 1, 2]
ser = pd.Series(np.arange(3.))
print(ser.loc[1])

# ser2의 index [1, 2, 3]
ser2 = pd.Series(np.arange(3.), index = [1, 2, 3])
print(ser2.loc[1])
```

```python
#프로퍼티 iloc
ser = pd.Series(np.arange(3.))
print(ser.iloc[1])

ser2=pd.Series(np.arange(3.), index=[1,2,3])
print(ser2.iloc[1])
print(ser2.iloc[:2])
print(ser2.loc[:2])
```

```python
#index가 지정되지 않은 경우 loc은 마지막 index 정보까지 포함, iloc은 마지막 index 정보는 미포함
ser = pd.Series(np.arange(3.))
print(ser.loc[:2])#[0,1,2] index의 정보 출력

ser = pd.Series(np.arange(3.))
print(ser.iloc[:2])#[0,1] index의 정보 출력
```

```python
#순위는 같은 값을 가진 값끼리 별도 처리
#method : 동점자 간 등수 철 ㅣ방법, 먼저 동점자들도 연속적으로 순위 계산
#default=average(동점자 등수의 평균), min(동점자 중 가장 낮은 등수), max(동점자 중 가장 높은 등수)
#first(동점과 상관없이 낮은 index 정보부터 순위), dense(min과 유사, 동점자 수와 상관없이 다음 등수는 1만 증가)

#na_option : {'keep', 'top', 'bottom'}
#keep(결측치 NaN는 index 정보 위치 그대로 유지), top(결측치 NaN은 상단 순위로 출력),bottom(결측치 NaN은 한단 순위로 출력)

ss = pd.Series([7, -5, 7, 4, 2, 0, 4])

print(ss.rank())
print(ss.rank(method='first'))
print(ss.rank(ascending=False, method='max'))
print(ss.rank(method='first').sort_values())
```
