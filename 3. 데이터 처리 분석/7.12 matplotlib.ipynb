pyplot 
- matplotlib을 다루기 위한 가장 기초 형태
- pyploy의 내장된 기능들은 다른 객체들에서도 유사하게 동작
- pyploy을 통해 객체 Figure, 객체 Axes 등을 생성할 수 있음
- import시, plt라는 약어로 주로 이용

figure
- 그래프를 포함하는 최상위 컨테이너
- 객체 Axes를 담고 있으며 최종적으로 pyplot으로 전달

axes
- 실제 그래프
- 그래프에 필요한 title, axis label, legend 등은 Axes로 입력

```python
#pyplot으로 그래프 그리기
import matplotlib.pyplot as plt
plt.plot()
```

```python
#pyplot으로 간단한 그래프 그리기 - 직선 그래프
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = np.arange(10)

plt.plot(data)
```

```python
#pyplot으로 간단한 그래프 그리기 - 난수 그래프
plt.plot(np.random.randn(100))
```

```python
#pyplot으로 간단한 그래프 그리기 - sin 곡선
x = np.linspace(0, np.pi * 2 , 201)

plt.plot(x, np.sin(x))
```

```python
#1000개의 난수의 누적합(cumsum) 그래프
randn_cumsum = np.random.randn(1000).cumsum()

#subplots는 matplotlib.pyplot의 가장 기초적인 메소드로 subplot들의 행/열 수를 나타내는 (nrows, ncols)를 매개변수로 가짐
#반환값은 fig, ax로 각각 객체 Figure와 객체 Axes / Axes들의 배열을 반환, 인스턴스 fig는 인스턴스 ax 관리
fig, ax = plt.subplots()

#그래프 크기 팁
# plt.rcParams['figure.figsize'] = [10, 6] or plt.rc('figure', figsize=(10,6))
#객체 Figure
# plt.figure(figsize=(10,6)) or fig, ax = plt.subplots(figsize=(10,6))

#ax 그래프의 제목 설정
ax.set_title('matplotlib plot practice')

#x축에 붙을 이름 설정
ax.set_xlabel('stages')

#y축에 붙을 이름 설정
ax.set_ylabel('Random Value')

#x축 눈금 설정
ax.set_xticks([0, 250, 500, 750, 1000])

#x축 눈금 라벨 설정, x_ticks를 설정하지 않으면 잘못된 라벨이 출력될 수 있음
#매개변수 rotation : 출력할 글자 각도 / fontsize : 출력할 글자 크기
#동일한 이름의 Y축 메소드도 있음
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')

#그래프 위에 문자열 표시하고 싶을 때 사용, 특이점(최대값/최소값/평균 위치)
#첫 번째 매개변수 : 출력하고자 하는 문자열 - 한글이 안될 수 있음
#매개변수 xy : 표시하고자 하는 X, Y 좌표
#매개변수 xytext : 문자열을 표시하고자 하는 X, Y좌표 - 아래 arrowprops 화살표 표시와 연동
#매개변수 arrowprops : 자료형 dict로 화살표 모양을 디자인
# => facecolor(화살표 색), headwidth(화살표 머리 두께), width(화살표 몸통 두께), headlength(활사표 머리 길이)
#매개변수 horizontalalignment : 왼쪽/가운데/오른쪽 문자열 정렬
#매개변수 verticalalignment : 상/중/하 문자열 정렬
ax.annotate('250\n250.00\n', xy=(250, randn_cumsum[250]+0.5),
            xytext=(250, randn_cumsum[250]+10),
            arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
            horizontalalignment='right', verticalalignment='top')

#메소드 set_xlim, set_ylim : 실제 출력할 X/Y축 데이터 범위(확대 개념)
# ax.set_xlim(0, 250)
# ax.set_ylim(-20, 30)

#메소드 plot : 선으로 출력할 데이터 입력
ax.plot(randn_cumsum)

#메소드 show : plot에 포함된 객체(인스턴스 fig)를 출력
plt.show()
```

```python
#실습 1 : 앞의 코드를 작성하고 다음 수정사항을 적용한다.
#ax 제목, X 축, Y 축 이름 변경
#X 축의 눈금 개수와 라벨 변경
#Y 축의 새로운 눈금과 라벨 생성
#별도로 표시한 250번쨰 지점에 대해 - 출력할 문자열 변경, 데이터 밑에 문자열 표시, 대각선 방향에 문자열 표시
randn_cumsum = np.random.randn(1000).cumsum()

fig, ax = plt.subplots()

plt.rcParams['figure.figsize'] = [10, 6]

ax.set_title('class practice 1')

ax.set_xlabel('x label')

ax.set_ylabel('y label')

ax.set_xticks([0, 200, 400, 600, 800, 1000])

ax.set_xticklabels(['0', '200', '400', '600', '800', '1000'], rotation=30, fontsize='small')

ax.set_yticks([-20, -10, 0, 10, 20])

ax.set_yticklabels(['F', 'E', 'D', 'C', 'B', 'A'], rotation=30, fontsize='small')

ax.annotate('x,y point', xy=(250, randn_cumsum[250]-0.5),
            xytext=(200, randn_cumsum[250]-10),
            arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
            horizontalalignment='right', verticalalignment='bottom')


ax.plot(randn_cumsum)

plt.show()
```

```python
#3개의 객체 Axes를 포함한 객체 Figure
fig = plt.figure()

#메소드 add_subplot : 인스턴스 fig 안에 인스턴스 axes를 생성
#매개변수 (int, int, index) => int:객체 fig 안에 생성할 객체 axes의 행/열, index:생성된 axes의 번호(격자형 구조에서 좌상단>우상단>좌하단 순)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

np_randn = np.random.randn(50)

#인스턴스 ax1 : 히스토그램(빈도수 확인 목적), bins(막대 두께), color(막대 색), alpha(막대 색 투명도)
ax1.hist(np_randn, bins=20, color='k', alpha=0.3)

#인스턴스 ax2 : 산포도 그래프, 매개변수 1,2번(X축 값, Y축 값)
ax2.scatter(np.arange(50), np.arange(50)+3*np_randn)

#인스턴스 ax3 : 실선 그래프, 매개변수 fmt(선의 색과 모양 : k(검정), --(점선)) 
ax3.plot(np_randn.cumsum(), 'k--') #plot - 꺾은선 그래프

plt.show()
```

```python
#6개의 객체 Axes를 포함한 객체 figure
np_randn = np.random.randn(50)

#2행 3열의 객체 Axes를 가진 객체 Figure 생성
fig, axes = plt.subplots(2, 3)

#Numpy의 Indexing 방식인 [,]으로 접근
axes[0,0].hist(np_randn, bins=20, color='k', alpha=0.3)
axes[0,2].scatter(np.arange(50), np.arange(50)+3*np_randn)
axes[1,1].plot(np_randn.cumsum(), 'k--')

plt.show()
```

```python
#for문으로 axes 관리

fig, axes = plt.subplots(2,2)

for i in range(2):
  for j in range(2):
    axes[i,j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)

#객체 figure 내 객체 Axes 사이 간격, 매개변수 wspace(가로 간격), hspace(세로 간격)
fig.subplots_adjust(wspace=0, hspace=0)

plt.show()
```

```python
#하나의 그래프에 여러 값 겹쳐서 그리기
#하나의 선 그래프 그릴 때 매개변수 : 1개(Y축 데이터), 2개(위치에 따라 종류 다름), 3개(X축, Y축, 선 종류 데이터)
#두 개 이상의 선 그래프를 그릴 때 : X축 데이터, Y축 데이터, 선 종류 순서를 반복
plt.plot(randn_cumsum, 'k--')
plt.show()
```

```python
#여러 객체 Axes 중 하나의 Axes만 겹쳐 그리기
ran1_cumsum = np.random.randn(500).cumsum()
ran2_cumsum = np.random.randn(500).cumsum()
ran3_cumsum = np.random.randn(500).cumsum()

fig, axes = plt.subplots(2,2)

axes[0,0].set_title('ran1_cumsum')
axes[0,1].set_title('ran2_cumsum')
axes[1,0].set_title('ran3_cumsum')
axes[1,1].set_title('ran_all_cumsum')

axes[0,0].plot(ran1_cumsum, 'k--')
axes[0,1].plot(ran2_cumsum, 'k--')
axes[1,0].plot(ran3_cumsum, 'k--')

axes[1,1].plot(ran1_cumsum, 'r--', label='ran1_cumsum')
axes[1,1].plot(ran2_cumsum, 'b-', label='ran2_cumsum')
axes[1,1].plot(ran3_cumsum, 'go-', label='ran3_cumsum')

#범례를 출력: 매개변수 loc(범례의 위치)
axes[0,0].legend()
axes[1,1].legend(loc='best')

ax.plot(randn_cumsum)

plt.show()
```

```python
#각각 X축 Y축 기준의 막대 그래프
np_randn = np.random.randn(7)
label_x = list()
for i in range(7):
  label_x.append(str(i))

fig, axes = plt.subplots(2,1)

axes[0].set_xticks(range(7))
axes[0].set_xticklabels(label_x)
axes[0].bar(range(7), np_randn, color='k', alpha=0.7)

#다른 방향의 막대 그래프
axes[1].set_yticks(range(7))
axes[1].set_yticklabels(label_x)
axes[1].barh(range(7), np_randn, color='k', alpha=0.7)

plt.show()
```

```python
#객체 Series와 그래프
ss = pd.Series(np.random.randn(10).cumsum(),
               index=np.arange(0, 100, 10))
ax = ss.plot()

plt.show()
```

```python
#객체 DataFrame과 그래프
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
ax = df.plot()

plt.show()
```

```python
#객체 Series와 막대 그래프
ss = pd.Series(np.random.randn(16), index=list('abcdefghijklmnop'))

fig, axes = plt.subplots(2,1)

ss.plot.bar(ax=axes[0], color='k', alpha=0.7)
ss.plot.barh(ax=axes[1], color='k', alpha=0.7)

axes[0].legend()
axes[1].legend()

plt.show()
```

```python
#객체 DataFrame과 막대 그래프
df = pd.DataFrame(np.random.rand(6,4), 
                  index=['one', 'two', 'three', 'four', 'five', 'size'],
                  columns=pd.Index(['A', 'B', 'C', 'D']))
ax = df.plot.bar()
ax.set_xlabel('Stages')

plt.show()
```

```python
#객체 DataFrame과 누적 막대 그래프, 매개변수 stacked=True(쌓다)
ax = df.plot.barh(stacked=True, alpha=0.5)
ax.set_ylabel('Value')

plt.show()
```
