```python
#DataFrame 형식으로 CSV파일 불러오기
import pandas as pd
#매개변수 encoding은 필요에 의해 사용
df_csv = pd.read_csv('/content/drive/My Drive/Colab Notebooks/데청캠2022/전국무인교통단속카메라표준데이터.csv', encoding='CP949')
print(df_csv.head(3))
print(df_csv.shape)#행/열 크기 확인
```

```python
#사본 DataFrame 생성 - 원본 데이터와 비교하기 위한 사본
df_new = pd.DataFrame(df_csv, copy=True) #혹은 df_new = df_csv.copy()
```

```python
#새로운 Column 생성
df_new['시도군구명'] = df_csv['시도명'].add(" ").add(df_csv['시군구명'])
```

```python
#겹치지 않는 "시도군구명" 열 값 출력
df_new['시도군구명'].unique
```

```python
#"진구" 문자열을 포함한 "시도군구명" 열 값 출력
for i in df_new["시도군구명"].unique():
  if "진구" in i:
    print(i)
```

목표 : 부산광역시진구를 부산진구로 통일

```python
#값이 일치하는 행 찾기
df_new['시군구명'] == '부산광역시진구'
#메소드 str.contains는 객체 Series가 가진 자료형이 str일 경우, 문자열 혹은 정규 표현식에 일치하는 열의 값만 반환
df_new[df_new['시군구명'].str.contains('진구')]
```

```python
#문자열을 포함하는 값을 가진 행 찾기
df_new['시군구명'].str.contains('진구')
df_new[df_new['시군구명'].str.contains('진구')]
```

```python
#"시군구명" 열에서 "부산광역시진구" 문자열과 일치하는 행만 새로운 DataFrame으로 분리
df_bjd = pd.DataFrame(df_new[df_new['시군구명']=="부산광역시진구"], copy = True)
```

```python
for i in df_bjd.columns[:3]:
  for j in df_bjd.index[:3]:
    print(df_bjd[i][j], end=", ")
  print()

df_bjd.head(3)
```

```python
#"부산광역시진구" 문자열을 "부산진구"로 변경
for i in df_bjd.columns:
  for j in df_bjd.index:
    if "부산광역시진구" in str(df_bjd[i][j]):
      #메소드 replace는 해당 DataFrame 행/열 전체 값을 탐색하여 매개변수 value로 교체
      df_bjd[i][j] = df_bjd[i][j].replace('부산광역시진구', '부산진구')
```

```python
#변경한 DataFrame을 원본 DataFrame에 덮어쓰기
df_final = pd.DataFrame(df_new, copy=True)
df_final.update(df_bjd)

df_final[df_final['시군구명']=='부산광역시진구']
```
